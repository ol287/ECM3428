```python
# Create and load the Karate Club graph, a well-known social network dataset included in NetworkX
karate = nx.karate_club_graph()
```

```python
# Calculate and return the average shortest path length of the Karate Club graph,
# representing the average number of steps required to travel between any two nodes in the graph
nx.average_shortest_path_length(karate)
```

```python
# Compute and return the shortest path between node 18 and node 16 in the Karate Club graph,
# represented as a list of nodes forming the path
nx.shortest_path(karate, 18, 16)
```

```python
# Compute and return the shortest path length between node 18 and node 16 in the Karate Club graph
# Total number of edges gone down
nx.shortest_path_length(karate, 18, 16)
```

```python
# Calculate and return the diameter of the Karate Club graph,
# which is the longest shortest path between any two nodes in the graph
nx.diameter(karate)
```

```python
# Calculate the diameter of the Karate Club graph, which is the longest shortest path
diameter = nx.diameter(karate)

# Generate a dictionary of shortest paths for all pairs of nodes in the graph
shortest_path_generator = nx.all_pairs_shortest_path(karate)

# Iterate through each node and its associated shortest path dictionary
for node, path_dictionary in shortest_path_generator:
    # Iterate through each pair of nodes and their shortest path
    for key, value in path_dictionary.items():
        # Check if the length of the path is equal to the graph's diameter
        if len(value) == diameter:
            # Print the path and its length
            print('Path:', value, '- Length:', len(value))
```

```python
# Apply the Girvan-Newman algorithm to detect communities in the Karate Club graph,
# which works by iteratively removing edges with the highest betweenness centrality to find community structure
nx.community.girvan_newman(karate)
```

```python
# Apply the Girvan-Newman algorithm to the Karate Club graph and convert the resulting community partitions into a list
# Each partition represents a set of nodes that belong to the same community
all_partitions = list(nx.community.girvan_newman(karate))
```

```python
# Access the first community partition from the list of partitions generated by the Girvan-Newman algorithm
# all_partitions[0] represents the initial division of nodes into communities after the first iteration of edge removal
all_partitions[0]
```

```python
# Access the second community partition from the list of partitions generated by the Girvan-Newman algorithm
# all_partitions[1] represents the division of nodes into communities after the second iteration of edge removal
all_partitions[1]
```

```python
# Access the third community partition from the list of partitions generated by the Girvan-Newman algorithm
# all_partitions[2] represents the division of nodes into communities after the second iteration of edge removal
all_partitions[2]
```

```python
# Calculate node positions using the Kamada-Kawai force-directed layout algorithm.
# The Kamada-Kawai algorithm is a spring-layout method where nodes are treated as repelling particles,
# and edges are modeled as springs that attract connected nodes. The algorithm positions nodes in a way that minimizes
# the energy of the graph by adjusting the node positions to make the graph's layout reflect the pairwise distances
# specified by the graph's edges, with the goal of achieving a visually balanced and interpretable layout.
node_layout = nx.kamada_kawai_layout(karate)
```

```python
# Access the third community partition from the list of partitions generated by the Girvan-Newman algorithm
# all_partitions[2] contains the nodes grouped into communities after the third iteration of edge removal
my_partition = all_partitions[2]

# Define a color palette with a set of distinct colors (tab20 palette from Seaborn)
# This palette will be used to assign different colors to each community in the graph
colours = sns.color_palette('tab20')

# Define the layout for node positions (e.g., Kamada-Kawai layout)
layout = nx.kamada_kawai_layout(karate)

# Iterate through each community in the partition and assign a color from the palette
# The 'zip' function pairs each community with a color from the 'colours' list
for community, colour in zip(my_partition, colours):
    # Draw nodes for each community with the corresponding color
    # 'nodelist' specifies the nodes to draw, 'node_color' assigns the community's color
    nx.draw_networkx_nodes(karate, layout, nodelist=community, node_color=colour)

# Draw edges of the graph using the Kamada-Kawai layout for node positioning
# 'width' controls the edge thickness, 'alpha' sets transparency for edge visibility
nx.draw_networkx_edges(karate, node_layout, width=1.0, alpha=0.5)

# Display the graph with the communities colored differently and edges drawn
plt.show()
```

```python
# Import the modularity function from NetworkX's community algorithms module.
# The modularity function is used to measure the strength of division of a network into communities.
# It compares the number of edges within communities to the number of edges expected in a random graph with the same degree distribution.
# A higher modularity value indicates a stronger community structure, with well-defined communities, while a value close to zero suggests a weak or no community structure.
from networkx.algorithms.community import modularity
```

```python
modularity_values = [ modularity(karate, partition) for partition in all_partitions ]

plt.figure(figsize=(15,4))
plt.plot(range(2, len(modularity_values)+2), modularity_values, '.-')
plt.xlabel('How many communities')
plt.ylabel('Modularity')
plt.xlim(0,35)
plt.grid()
plt.show()
```
